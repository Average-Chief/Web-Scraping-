```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url = 'https://www.imdb.com/list/ls521050493/?ref_=fea_c_cseghw_50dplus_cta'
page = requests.get(url)
soup = bs(page.text,'html')
table = soup.find('div', class_ = 'lister-list')
print(table)
```

```python
movies = table.find_all('div', class_='lister-item-content')
for index, movie in enumerate(movies, start=1):
    title_element = movie.find('a') 
    rating_element = movie.find('strong')
    
    if title_element:
        title = title_element.text 
        print(f"{index}. {title}")

    if rating_element:
        rating = rating_element.text.strip()
        print(f"Rating: {rating}")
```
